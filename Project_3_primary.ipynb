```python
df=pd.read_csv('Dataset - Project 3.csv')
```

```python
df['Transaction_Frequency'] = df.groupby('Customer_ID')['Transaction_ID'].transform('count')
```

New column for transaction frequency


```python
df
```

```python
average_spending = df.groupby('Customer_ID')['Transaction_Amount'].transform('mean')
df['Deviation_Avg_Spending'] = df['Transaction_Amount'] - average_spending
```

Calculating avg deviation from spending

```python
df
```

```python
threshold = 8000
df['Anomaly_Threshold'] = df['Transaction_Amount'] > threshold
```

Anomaly calculation based on threshold defined

```python
mean = np.mean(df['Transaction_Amount'])
std = np.std(df['Transaction_Amount'])
df['Z-Score'] = (df['Transaction_Amount'] - mean) / std
z_threshold = 2
# Common threshold for z-scores is 2

df['Anomaly_Z-Score'] = df['Z-Score'].abs() > z_threshold
```
